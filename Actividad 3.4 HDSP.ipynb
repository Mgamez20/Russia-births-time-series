## Autoregressive Model

```python
# Visualización de datos originales
plt.subplots(figsize = (15,5))
plt.plot(data_f)
plt.title('Nacimientos en Rusia 1959-2014 (Serie Original)', Fontsize = 15)
plt.xlabel('Año')
plt.ylabel('Nacimientos (Millones)')
plt.show()
```

```python
# Determinar si la serie es estacionaria
from statsmodels.tsa.stattools import adfuller

dftest = adfuller(data_f, autolag = 'AIC')

print('P-valor: ', dftest[1])
print('P-valor > 0.05')
```

```python
# Conversión de la serie a estacionaria
data_f_stationary = data_f.diff().dropna()
plt.subplots(figsize = (15,5))
plt.plot(data_f_stationary)
plt.title('Nacimientos en Rusia 1959-2014 (Serie Estacionaria)', Fontsize = 15)
plt.show()

dftest = adfuller(data_f_stationary, autolag = 'AIC')

print('P-valor (estacionaria): ', dftest[1])
print('P-valor < 0.05')
```

```python
#Creación del modelo Autoregresivo
from statsmodels.tsa.ar_model import AutoReg

train, test = data_f_stationary[:45], data_f_stationary[45:]

AutoReg = AutoReg(train, lags = 4)
AutoReg_model = AutoReg.fit()
AutoReg_model.summary()
```

```python
# Predicción del modelo (10 Años)
start = len(train) 
end = start + len(test) - 1 
pred = AutoReg_model.predict(start = start, end = end )
pred
```

```python
#Cambio de index predicciones (num -> año)
pred.index = pd.date_range(start = '2005', end = '2015', freq = 'A').year
pred
```

```python
#Visualización de Datos 
plt.subplots(figsize = (15,5))
plt.plot(train, color = 'black', marker = '.', label = 'Serie Original Train (Estacionaria)')
plt.plot(test, color = 'red', marker = '.', label = 'Serie Original Test (Estacionaria)')
plt.plot(pred, color = 'blue', marker = '.', label = 'Predicción por modelo')
plt.title('Modelo Autoregresivo en Nacimientos de Rusia 1959-2014 (Serie Estacionaria)', Fontsize = 15)
plt.xlabel('Año', Fontsize = 10)
plt.legend()
plt.show()
```

```python
#Cálculo de RMSE y MAE
import math
print('Root Mean Squared Error: {0:.2f}'.format(math.sqrt(mean_squared_error(test, pred))))
print('Mean Absolute Error: {0:.2f}'.format(mean_absolute_error(test, pred)))
```
